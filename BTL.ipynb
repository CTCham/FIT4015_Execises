```python
import tkinter as tk
from tkinter import ttk, messagebox, PhotoImage
import json
import os

FILE_NAME = "sbt.json"

def load_data_to_table():
    for row in tree.get_children():
        tree.delete(row)
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r", encoding="utf-8") as f:
            try:
                ds = json.load(f)
                for item in ds:
                    tree.insert("", "end", values=(
                        item["ma_sv"], item["ho_ten"], item["lop"], item["so_bai_da_hoan_thanh"], item["diem"]
                    ))
            except:
                pass

def save_data():
    data = {
        "ma_sv": entry_ma_sv.get(),
        "ho_ten": entry_ho_ten.get(),
        "lop": entry_lop.get(),
        "so_bai_da_hoan_thanh": entry_so_bai.get(),
        "diem": entry_diem.get()
    }
    if not all(data.values()):
        messagebox.showwarning("Thiếu thông tin", "Vui lòng nhập đầy đủ thông tin!")
        return

    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r", encoding="utf-8") as f:
            try:
                ds = json.load(f)
            except:
                ds = []
    else:
        ds = []

    ds.append(data)
    with open(FILE_NAME, "w", encoding="utf-8") as f:
        json.dump(ds, f, ensure_ascii=False, indent=4)
    messagebox.showinfo("Thành công", "Đã lưu thông tin bài tập!")
    clear_entries()
    load_data_to_table()

def clear_entries():
    entry_ma_sv.delete(0, tk.END)
    entry_ho_ten.delete(0, tk.END)
    entry_lop.delete(0, tk.END)
    entry_so_bai.delete(0, tk.END)
    entry_diem.delete(0, tk.END)

def exit_form():
    root.destroy()

def about():
    messagebox.showinfo("Thông tin", "Ứng dụng quản lý sách bài tập và lịch ôn tập\nTác giả: Bạn")

root = tk.Tk()
root.title("Quản Lý Sách Bài Tập & Lịch Ôn Tập")
root.geometry("1400x800")
root.configure(bg="#e3f2fd")

# Menu
menu_bar = tk.Menu(root)

menu_baitap = tk.Menu(menu_bar, tearoff=0)
menu_baitap.add_command(label="Thêm mới", command=save_data)
menu_baitap.add_command(label="Xem danh sách", command=load_data_to_table)
menu_baitap.add_command(label="Sửa bài tập", command=None)
menu_baitap.add_command(label="Xóa bài tập", command=None)
menu_baitap.add_separator()
menu_baitap.add_command(label="Đánh dấu hoàn thành", command=None)
menu_baitap.add_command(label="Xem chi tiết", command=None)
menu_baitap.add_separator()
menu_baitap.add_command(label="Thoát", command=exit_form)
menu_bar.add_cascade(label="Quản lý bài tập", menu=menu_baitap)

menu_lich = tk.Menu(menu_bar, tearoff=0)
menu_lich.add_command(label="Xem lịch", command=None)
menu_lich.add_command(label="Thêm lịch", command=None)
menu_lich.add_command(label="Thống kê tổng số bài", command=None)
menu_lich.add_command(label="Thống kê điểm trung bình", command=None)
menu_bar.add_cascade(label="Lịch ôn tập", menu=menu_lich)

menu_file = tk.Menu(menu_bar, tearoff=0)
menu_file.add_command(label="Xuất Excel", command=None)
menu_file.add_command(label="Xuất PDF", command=None)
menu_file.add_command(label="Nhập từ file", command=None)
menu_file.add_command(label="In danh sách", command=None)
menu_file.add_separator()
menu_file.add_command(label="Đặt lại dữ liệu", command=None)
menu_bar.add_cascade(label="Tệp", menu=menu_file)

menu_account = tk.Menu(menu_bar, tearoff=0)
menu_account.add_command(label="Đổi giao diện", command=None)
menu_account.add_command(label="Đổi mật khẩu", command=None)
menu_account.add_command(label="Đăng xuất", command=None)
menu_bar.add_cascade(label="Tài khoản", menu=menu_account)

menu_help = tk.Menu(menu_bar, tearoff=0)
menu_help.add_command(label="Trợ giúp", command=about)
menu_help.add_command(label="Giới thiệu", command=about)
menu_bar.add_cascade(label="Trợ giúp", menu=menu_help)

root.config(menu=menu_bar)

# Thanh chức năng ngang (toolbar)
toolbar = tk.Frame(root, bg="#f5f5f5", height=36)
toolbar.pack(side="top", fill="x", pady=(0, 2))

button_style = {
    "font": ("Arial", 10, "bold"),
    "bg": "#1976d2",
    "fg": "white",
    "width": 10,
    "height": 1,
    "relief": "groove",
    "bd": 1
}

tk.Button(toolbar, text="Thêm mới", command=save_data, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Sửa", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Xóa", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Làm mới", command=load_data_to_table, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Tìm kiếm", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Lọc theo lớp", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Lọc theo điểm", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Xuất Excel", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Xuất PDF", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Nhập từ file", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Thống kê", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Đặt lại dữ liệu", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Đánh dấu hoàn thành", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Xem chi tiết", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="In danh sách", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Đổi giao diện", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Đổi mật khẩu", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Đăng xuất", command=None, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Trợ giúp", command=about, **button_style).pack(side="left", padx=3, pady=3)
tk.Button(toolbar, text="Giới thiệu", command=about, **button_style).pack(side="left", padx=3, pady=3)

# Main layout
main_frame = tk.Frame(root, bg="#e3f2fd")
main_frame.pack(fill="both", expand=True)

# Khung trái (tổng quan)
left_frame = tk.Frame(main_frame, bg="#e1f5fe", width=350, bd=2, relief="groove")
left_frame.pack(side="left", fill="y", padx=10, pady=10)
tk.Label(left_frame, text="Tổng quan bài tập", bg="#e1f5fe", fg="#1976d2", font=("Arial", 15, "bold")).pack(pady=20)
tk.Label(left_frame, text="Nhập thông tin bài tập mới:", bg="#e1f5fe", fg="#1976d2", font=("Arial", 12, "bold")).pack(pady=10)

form_inner = tk.Frame(left_frame, bg="#e1f5fe")
form_inner.pack(pady=5)

tk.Label(form_inner, text="Mã sinh viên:", bg="#e1f5fe", fg="#1976d2", font=("Arial", 11)).grid(row=0, column=0, sticky="e", pady=5)
entry_ma_sv = tk.Entry(form_inner, font=("Arial", 11), width=22)
entry_ma_sv.grid(row=0, column=1, pady=5)
tk.Label(form_inner, text="Họ tên:", bg="#e1f5fe", fg="#1976d2", font=("Arial", 11)).grid(row=1, column=0, sticky="e", pady=5)
entry_ho_ten = tk.Entry(form_inner, font=("Arial", 11), width=22)
entry_ho_ten.grid(row=1, column=1, pady=5)
tk.Label(form_inner, text="Lớp:", bg="#e1f5fe", fg="#1976d2", font=("Arial", 11)).grid(row=2, column=0, sticky="e", pady=5)
entry_lop = tk.Entry(form_inner, font=("Arial", 11), width=22)
entry_lop.grid(row=2, column=1, pady=5)
tk.Label(form_inner, text="Số bài đã hoàn thành:", bg="#e1f5fe", fg="#1976d2", font=("Arial", 11)).grid(row=3, column=0, sticky="e", pady=5)
entry_so_bai = tk.Entry(form_inner, font=("Arial", 11), width=22)
entry_so_bai.grid(row=3, column=1, pady=5)
tk.Label(form_inner, text="Điểm:", bg="#e1f5fe", fg="#1976d2", font=("Arial", 11)).grid(row=4, column=0, sticky="e", pady=5)
entry_diem = tk.Entry(form_inner, font=("Arial", 11), width=22)
entry_diem.grid(row=4, column=1, pady=5)

# Khung phải (bảng dữ liệu)
right_frame = tk.Frame(main_frame, bg="#e3f2fd")
right_frame.pack(side="left", fill="both", expand=True, padx=10, pady=10)

# Thanh tìm kiếm
search_frame = tk.Frame(right_frame, bg="#e3f2fd")
search_frame.pack(fill="x", pady=5)
tk.Label(search_frame, text="Tìm kiếm:", bg="#e3f2fd", font=("Arial", 11)).pack(side="left", padx=5)
search_var = tk.StringVar()
search_entry = tk.Entry(search_frame, textvariable=search_var, font=("Arial", 11), width=30)
search_entry.pack(side="left", padx=5)

def search_data(*args):
    keyword = search_var.get().lower()
    for row in tree.get_children():
        tree.delete(row)
    if os.path.exists(FILE_NAME):
        with open(FILE_NAME, "r", encoding="utf-8") as f:
            try:
                ds = json.load(f)
                for item in ds:
                    if (keyword in item["ma_sv"].lower() or
                        keyword in item["ho_ten"].lower() or
                        keyword in item["lop"].lower()):
                        tree.insert("", "end", values=(
                            item["ma_sv"], item["ho_ten"], item["lop"], item["so_bai_da_hoan_thanh"], item["diem"]
                        ))
            except:
                pass

search_var.trace("w", search_data)

# Bảng dữ liệu
columns = ("ma_sv", "ho_ten", "lop", "so_bai_da_hoan_thanh", "diem")
tree = ttk.Treeview(right_frame, columns=columns, show="headings", height=20)
tree.heading("ma_sv", text="Mã SV")
tree.heading("ho_ten", text="Họ tên")
tree.heading("lop", text="Lớp")
tree.heading("so_bai_da_hoan_thanh", text="Số bài")
tree.heading("diem", text="Điểm")
tree.pack(fill="both", expand=True, pady=10)

# Tùy chọn màu sắc cho bảng
style = ttk.Style()
style.theme_use("default")
style.configure("Treeview", background="#e3f2fd", fieldbackground="#e3f2fd", font=("Arial", 11))
style.configure("Treeview.Heading", font=("Arial", 12, "bold"), background="#1976d2", foreground="white")

load_data_to_table()

root.mainloop()
```
